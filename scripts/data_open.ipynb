```python
from collections import Counter
import re
```

```python
all_text = " ".join(df["리뷰"].astype(str))
words = re.findall(r'\b[가-힣]+\b', all_text)
word_freq = Counter(words)

top_n = word_freq.most_common(1000)
ranking_df = pd.DataFrame(top_n, columns=["단어", "빈도수"])

print(ranking_df)
```

# 속성 키워드 랭킹1000에서 추출

```python
# 속성 키워드
keywords = {
    "머릿결": ["머릿결", "머리결"],
    "향": ["향","냄새"],
    "거품": ["거품", "폼"],
    "세정력": ["세정","씻기"],
    "자극도": ["자극","순하","순해"],
    "탈모": ["탈모"],
    "쿨링": ["쿨","화한","열이"],
    "가격": ["가격","가성비","저렴"],
    "염색": ["염색"],
    "보색": ["보색"]
}
```

```python
result = []

for keyword, List in keywords.items():
    filtered = ranking_df[ranking_df["단어"].apply(lambda x: any(s in x for s in List))]
    wordlist = filtered["단어"].tolist()
    freq = filtered["빈도수"].sum()
    
    result.append({
        "키워드": keyword,
        "단어목록": wordlist,
        "빈도합": freq
    })


result_df = pd.DataFrame(result)
result_df.set_index("키워드", inplace=True)
result_df
```
